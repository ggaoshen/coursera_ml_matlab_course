```python
# Visualizing Data
def unpack_serialized_pixel(array,dim=(20,20)):
    # 1*400 array represents 20*20 pixel picture in grey scale
    return np.reshape(array,dim).T

def unpack_multiple_pixel_arrays(df, dim=(20,20)):
    input_matrix=df.values
    n_images=input_matrix.shape[0]
    print(n_images)
    n_cols=int(np.floor(np.sqrt(n_images)))
    if (1.*n_images)%n_cols==0: n_rows=int(np.floor(1.*n_images/n_cols))
    else: n_rows=int(np.floor(1.*n_images/n_cols))+1
    # print(n_cols, n_rows)
    output_block=np.zeros((n_rows*20, n_cols*20))
#     print(output_block.shape)
    # agg_array=[]
    for i in range(n_images):

        block=unpack_serialized_pixel(input_matrix[i,:])
    #     print(block.shape)
        col=i%n_cols
        row=int(np.floor(i/n_cols))
    #     print(row, col)
    #     print(output_block[row*20:(row+1)*20,col*20:(col+1)*20].shape)
        output_block[row*20:(row+1)*20,col*20:(col+1)*20]=block
    return output_block
```

```python
# 1.3 Vectorizing Logistic Regression

# 1.3.1 Vectorize cost function

# logistic cost function
def logit_fn(z): # logistic(x) - point calculation
    return 1./(1+np.exp(-z))
    
def regression_fn(thetas,x_matrix): # h(x)
    return logit_fn(x_matrix.dot(thetas)) # X*theta

def logistic_cost_fn(thetas, x_matrix, y_vector): # J(theta)
    m=len(y_vector)
    log_h=np.log(regression_fn(thetas,x_matrix)) # vector of dim = n_observation
    log_1mh=np.log(1-regression_fn(thetas,x_matrix)) # vector of dim = n_observation
    return 1./m*(-np.inner(y_vector,log_h)-np.inner((1-y_vector), log_1mh))


# 1.3.2 Vectorize gradient
def gradient(thetas, x_matrix, y_vector): # return all gradients
    m=len(y_vector)
    gradients=[]
    log_h=regression_fn(thetas,x_matrix)
    return 1./m*x_matrix.T.dot(log_h-y) # J'(theta_i)=1/m * X^T *(h(theta_i)-y)
```

```python
# 1.3.3 Vectorize Regularized logistic regression cost function and gradient
def reg_cost_fn(thetas, lamb, x_matrix, y_vector): # J(theta) with penalty terms
    m=len(y_vector)
    log_h=np.log(regression_fn(thetas,x_matrix)) # vector of dim = n_observation
    log_1mh=np.log(1-regression_fn(thetas,x_matrix)) # vector of dim = n_observation
    result_unreg=1./m*(-np.inner(y_vector,log_h)-np.inner((1-y_vector), log_1mh))
    thetas_unpenalize1st=thetas.copy()
    thetas_unpenalize1st[0]=0. # NOT PENALIZING THETA0
    return result_unreg+lamb/(2.*m)*np.sum(thetas_unpenalize1st**2)

# 1.3.2 Vectorize gradient (regularized)
def reg_gradient(thetas, lamb, x_matrix, y_vector): # return all gradients
    m=len(y_vector)
    gradients=[]
    log_h=regression_fn(thetas,x_matrix)
#     print((y_vector).shape)
#     print(1./m*x_matrix.T.dot(log_h-y_vector)+lamb/m*thetas)
    result_unreg=1./m*x_matrix.T.dot(log_h-y_vector)
    thetas_unpenalize1st=thetas.copy()
    thetas_unpenalize1st[0]=0. # NOT PENALIZING THETA0
    return result_unreg+lamb/m*thetas_unpenalize1st # J'(theta vector)=1/m * X^T *(h(X)-y) + lambda/m*thetas
```

```python
# 1.4 One-vs-all classification
X_aug=X.copy()
X_aug[-1]=1 # pad X matrix with column of 1s
X_aug=X_aug.sort_index(axis=1) # reindex such that the constant column becomes the first column
# print(X_aug)
thetas_0=np.array([0.01]*X_aug.shape[1]) # theta0 initialized to be all 0.01s - this won't have any numerical issues 
lamb=.1 # initialize regularization parameter

categories=y.drop_duplicates().values
model_array=[]
for i in categories: # test with only one category
    y_binary=np.squeeze(y==i) # 0 is represented as 10 in y, so no conversion is necessary
    opt_cost_fn=lambda params: reg_cost_fn(params, lamb, X_aug, y_binary) # y_binary is 0,1 for 1 category
    opt_grad_fn=lambda params: reg_gradient(params, lamb, X_aug, y_binary)#!!!!!!!!!!!
    res = optimize.fmin_cg(f=opt_cost_fn, x0=thetas_0, fprime=opt_grad_fn)
#     print(res)
    model_array.append(res)
model_array=np.array(model_array) # trained model result (10 linear models, one for each number)
```
