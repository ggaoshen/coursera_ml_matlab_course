```python
# Visualizing Data
def unpack_serialized_pixel(array,dim=(20,20)):
    # 1*400 array represents 20*20 pixel picture in grey scale
    return array.values.reshape(dim).T

def unpack_multiple_pixel_arrays(df, dim=(20,20)):
    input_matrix=df
    n_images=input_matrix.shape[0]
#     print(n_images)
    n_cols=int(np.floor(np.sqrt(n_images)))
    if (1.*n_images)%n_cols==0: n_rows=int(np.floor(1.*n_images/n_cols))
    else: n_rows=int(np.floor(1.*n_images/n_cols))+1
#     print(n_cols, n_rows)
    output_block=np.zeros((n_rows*20, n_cols*20))
#     print(output_block.shape)
    # # agg_array=[]
    for i in range(n_images):
        block=unpack_serialized_pixel(input_matrix.iloc[i,:])
    #     print(block.shape)
        col=i%n_cols
        row=int(np.floor(i/n_cols))
    #     print(row, col)
    #     print(output_block[row*20:(row+1)*20,col*20:(col+1)*20].shape)
        output_block[row*20:(row+1)*20,col*20:(col+1)*20]=block
    return output_block

def displayData(df, dim=(20,20)):
    out = unpack_multiple_pixel_arrays(df)
    plt.figure(num=None, figsize=(8, 6))
    plt.imshow(out, cmap='gray')
    plt.show()
```

```python
# logistic function
def logit_fn(z): # logistic(x) - point calculation
    return 1./(1+np.exp(-z))
```

```python
# forward propagation
def predict(X_test, thetas): 
    # X_test size 5000*401
    # thetas=[theta1, theta2]
    
    nnweights_theta_1=thetas[0]
    nnweights_theta_2=thetas[1]
    
    X_test.insert(loc=0, column="0", value=np.ones(X_test.shape[0])) # pad X_test with column of 1s on front
    X_test.columns=range(X_test.shape[1]) # rename X_test col names to be a sequence
    
    mid_values=logit_fn(X_test.dot(nnweights_theta_1.T)) # mid_values size 5000*25
#     print(mid_values)
    mid_values.insert(loc=0, column="0", value=np.ones(mid_values.shape[0])) # pad X_test with column of 1s on front
    mid_values.columns=range(mid_values.shape[1]) # name columns from 1 - 10 so that matches with weights definition
    
    result = logit_fn(mid_values.dot(nnweights_theta_2.T))
    result.columns=range(1,11)
    
    return result

thetas=[nnweights_theta_1, nnweights_theta_2]
prob_grid = predict(X.copy(), thetas)

prob_grid.ix[:,'PredValue']=prob_grid.columns[np.argmax(prob_grid.values,axis=1)] 
prob_grid
```

```python
# 2.1 Sigmoid gradient
def logit_gradient(z):
    gz=logit_fn(z)
    return gz*(1-gz)

# test z=0, gradient=0.25
print('z=0, logit_gradient=', logit_gradient(0))
# test z=9999, gradient=0
print('z=9999, logit_gradient=', logit_gradient(9999))

# function correct
```

```python
# 1.3 feedforward and cost function + 2.5 backprop
def nn_cost_function_serialized(X, y, num_input_layer, num_hidden_layer, num_labels, unrolled_thetas):
    # thetas: serialized theta: 1d
    # X: ndarray, y: ndarray
    
    theta1=np.reshape(unrolled_thetas[:num_hidden_layer*(num_input_layer+1)], (num_hidden_layer, num_input_layer+1))
    theta2=np.reshape(unrolled_thetas[num_hidden_layer*(num_input_layer+1):], (num_labels, num_hidden_layer+1))
    thetas=[theta1, theta2]
    m=X.shape[0]
    a=X.copy()
    zs_=[] # stores z2,z3!!
    as_=[] # stores a1,a2,a3
#     penalty=[]
    
    for theta in thetas:
        # pad a with zeros
        a_=np.ones((a.shape[0], a.shape[1]+1))
        a_[:, 1:]=a
        
        as_.append(a_) # appending a1=X, padded with 1s
        zs_.append(theta.dot(a_.T).T) # calc z2=theta1*a1'
        
        # calc next activation
        a=logit_fn(zs_[-1]) # 5000*25
        # calc squared penalty terms
#         penalty+=np.ones(theta.shape[0]).dot(theta[:,1:]**2).dot(np.ones(theta[:,1:].shape[1])) 
        # NOTE DON'T PENALIZE BIAS TERM (COSTANT)
    
    as_.append(a) # appending a3
    
    # make y a vector for each training set
    y_vec=np.zeros(a.shape)
    for i in range(y_vec.shape[0]):
        if y[i]==10: y_vec[i, 9]=1
        else: y_vec[i, y[i]-1]=1

    # calc cost
    log_a=np.log(a)
    log_1ma=np.log(1-a)
    cost=-(np.trace(y_vec.dot(log_a.T))+np.trace((1-y_vec).dot(log_1ma.T)))/m
    
    # calc gradient with backprop
    a1=as_[0]
    a2=as_[1]
    a3=as_[2]
    
    z2=zs_[0]
    z3=zs_[1]
    
    
    delta3 = a3 - y_vec # initilize delta as last layer difference 
    delta2 = delta3.dot(theta2)[:, 1:] * logit_gradient(z2)

    grad_theta2 = delta3.T.dot(a2)/m
    grad_theta1 = delta2.T.dot(a1)/m
#     grad_theta2 = delta3.T.dot(a2)[:,1:]/m
#     print(delta2.shape, a1.shape)
#     grad_theta1 = delta2.T.dot(a1)[:,1:]/m

    grad_unrolled = unroll_thetas([grad_theta1, grad_theta2])
    return cost, grad_unrolled

def unroll_thetas(thetas):
    res=np.array([])
    for theta in thetas:
        if isinstance(theta, pd.DataFrame):
            res = np.concatenate((res, theta.values.flatten()))
        else: res = np.concatenate((res, theta.flatten()))
    return res

unrolled_thetas=unroll_thetas([nnweights_theta_1, nnweights_theta_2])
nn_cost_function_serialized(X.values, y.values, X.shape[1], nnweights_theta_1.shape[0], nnweights_theta_2.shape[0], unrolled_thetas) 
# RESULT 0.2876295 -- CORRECT

```

```python
# 1.4 Regularized cost function + 2.5 backprop on regularized cost function

def nn_cost_function_reg(X, y, thetas, lmda=1):
    # thetas: list of weight matrices, from first layer to last layer 
    a=X.copy()
    m=X.shape[0] # number of observations
    penalty=0
    for theta in thetas:
        a.insert(loc=0, column='0', value=np.ones(a.shape[0]))# pad 0 vector to X
        a.columns=range(a.shape[1])
        a=logit_fn(theta.dot(a.T).T) # 5000*25
    
        # calc squared penalty terms
        penalty+=np.ones(theta.shape[0]).dot(theta.ix[:,1:].values**2).dot(np.ones(theta.ix[:,1:].shape[1])) 
        # NOTE DON'T PENALIZE BIAS TERM (COSTANT)

    # make y a vector for each training set
    y_vec=pd.DataFrame(np.zeros(a.shape)) 
    y_vec.columns=list(range(1,11))
    for i in range(y_vec.shape[0]):
        y_vec.ix[i, y.iloc[i]]=1

    log_a=np.log(a)
    log_1ma=np.log(1-a)
    cost=-(np.trace(y_vec.values.dot(log_a.T.values))+np.trace((1-y_vec).dot(log_1ma.T.values)))/m + penalty*lmda/(2*m)
    
    return cost

nn_cost_function_reg(X, y, [nnweights_theta_1, nnweights_theta_2]) # cost # compared to 0.383770 -- CORRECT
```

```python
# 1.4 Regularized cost function + 2.5 backprop on regularized cost function
def nn_cost_function_REG_serialized(X, y, num_input_layer, num_hidden_layer, num_labels, unrolled_thetas, lmda=1):
    # thetas: serialized theta: 1d
    # X: ndarray, y: ndarray
    
    theta1=np.reshape(unrolled_thetas[:num_hidden_layer*(num_input_layer+1)], (num_hidden_layer, num_input_layer+1))
    theta2=np.reshape(unrolled_thetas[num_hidden_layer*(num_input_layer+1):], (num_labels, num_hidden_layer+1))
    thetas=[theta1, theta2]
    m=X.shape[0]
    a=X.copy()
    zs_=[] # stores z2,z3!!
    as_=[] # stores a1,a2,a3
    penalty=0
    
    for theta in thetas:
        # pad a with zeros
        a_=np.ones((a.shape[0], a.shape[1]+1))
        a_[:, 1:]=a
        
        as_.append(a_) # appending a1=X, padded with 1s
        zs_.append(theta.dot(a_.T).T) # calc z2=theta1*a1'
        
        # calc next activation
        a=logit_fn(zs_[-1]) # 5000*25
        # calc squared penalty terms
        penalty+=np.ones(theta.shape[0]).dot(theta[:,1:]**2).dot(np.ones(theta[:,1:].shape[1])) 
        # NOTE DON'T PENALIZE BIAS TERM (COSTANT)
    
    as_.append(a) # appending a3
    
    # make y a vector for each training set
    y_vec=np.zeros(a.shape)
    for i in range(y_vec.shape[0]):
        if y[i]==10: y_vec[i, 9]=1
        else: y_vec[i, y[i]-1]=1

    # calc cost
    log_a=np.log(a)
    log_1ma=np.log(1-a)
    cost=-(np.trace(y_vec.dot(log_a.T))+np.trace((1-y_vec).dot(log_1ma.T)))/m + penalty*lmda/(2*m)
    
    # calc gradient with backprop
    a1=as_[0]
    a2=as_[1]
    a3=as_[2]
    
    z2=zs_[0]
    z3=zs_[1]
    
    # calc delta difference for each layer
    delta3 = a3 - y_vec # initilize delta as last layer difference 
    delta2 = delta3.dot(theta2)[:, 1:] * logit_gradient(z2)
    
    # removing penalty for bias term
    theta1_penalty=np.zeros(theta1.shape)
    theta1_penalty[:,1:]=theta1[:, 1:]
    theta2_penalty=np.zeros(theta2.shape)
    theta2_penalty[:,1:]=theta2[:, 1:]
    
    # calc gradient + penalty terms
    grad_theta2 = delta3.T.dot(a2)/m + lmda/m*theta2_penalty
    grad_theta1 = delta2.T.dot(a1)/m + lmda/m*theta1_penalty
#     grad_theta2 = delta3.T.dot(a2)[:,1:]/m
#     print(delta2.shape, a1.shape)
#     grad_theta1 = delta2.T.dot(a1)[:,1:]/m

    grad_unrolled = unroll_thetas([grad_theta1, grad_theta2])
    return cost, grad_unrolled

def unroll_thetas(thetas):
    res=np.array([])
    for theta in thetas:
        if isinstance(theta, pd.DataFrame):
            res = np.concatenate((res, theta.values.flatten()))
        else: res = np.concatenate((res, theta.flatten()))
    return res

unrolled_thetas=unroll_thetas([nnweights_theta_1, nnweights_theta_2])
nn_cost_function_REG_serialized(X.values, y.values, X.shape[1], nnweights_theta_1.shape[0], nnweights_theta_2.shape[0], unrolled_thetas) 
# cost # compared to 0.383770 -- CORRECT

```

```python
def numericalGradient(X_sample, y_sample, unrolled_thetas, ind_reg=0, eps=0.0001):
    # calculate gradient at i-th element in thetas  
    res=[]
    for i in range(len(unrolled_thetas)):
        if ind_reg==0: # without regularization
            unrolled_thetas[i]+=eps
            cost_plus, gd_plus = nn_cost_function_serialized(X_sample, y_sample, X_sample.shape[1], nnweights_theta_1.shape[0], nnweights_theta_2.shape[0], unrolled_thetas)
            unrolled_thetas[i]-=2*eps
            cost_minus, gd_minus = nn_cost_function_serialized(X_sample, y_sample, X_sample.shape[1], nnweights_theta_1.shape[0], nnweights_theta_2.shape[0], unrolled_thetas)
            res.append((cost_plus-cost_minus)/(2*eps))
        else: # with regularization
            unrolled_thetas[i]+=eps
            cost_plus, gd_plus = nn_cost_function_REG_serialized(X_sample, y_sample, X_sample.shape[1], nnweights_theta_1.shape[0], nnweights_theta_2.shape[0], unrolled_thetas)
            unrolled_thetas[i]-=2*eps
            cost_minus, gd_minus = nn_cost_function_REG_serialized(X_sample, y_sample, X_sample.shape[1], nnweights_theta_1.shape[0], nnweights_theta_2.shape[0], unrolled_thetas)
            res.append((cost_plus-cost_minus)/(2*eps))
            
    return res
```
